## Imports

```python
# Python Modules
# --------------------------------------------------
import os
import sys
import numpy             as np
import skimage.io        as io
import matplotlib.pyplot as plt



# Third Party Modules
# --------------------------------------------------
from pycocotools.coco import COCO


# Custom Modules
# --------------------------------------------------
import config
import utils


# Settings
# --------------------------------------------------
sys.path.append(config.COCO_API_PATH)
np.random.seed(config.RANDOM_SEED

%matplotlib inline
```

## Initialize COCO

```python
coco      = COCO(config.ANN_INSTANCE_VAL_FILE)
coco_caps = COCO(config.ANN_CAPTIONS_VAL_FILE)
image_ids = list(coco.anns.keys())
```

## Visualize Image

```python
# Get Random Data
# --------------------------------------------------
ann_id = np.random.choice(image_ids)
img_id = coco.anns[ann_id][config.COCO_IMG_ID_COL]

img = coco.loadImgs(img_id)[0]
url = img[config.COCO_URL_COL]
I   = io.imread(url)


# Visualize Image
# -------------------------------------------------- 
print(url)

plt.figure(figsize = (10, 10))
plt.axis("off")
plt.imshow(I)
plt.show()


# Visualize Caption
# -------------------------------------------------- 
annIds = coco_caps.getAnnIds(imgIds = img[config.IMG_UID_COL]);
anns   = coco_caps.loadAnns(annIds)

coco_caps.showAnns(anns)
```
